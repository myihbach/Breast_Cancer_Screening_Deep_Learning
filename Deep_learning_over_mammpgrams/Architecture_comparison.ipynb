```python
# import the architecture
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def load_histories(model):
    path = "/content/drive/MyDrive/Breast_Cancer/results/vectors/"+model+"/histories__vf.npy"
    return np.load(path,allow_pickle=True)

def save_avg(data ,name,model):
    data = pd.DataFrame(data).mean(axis = 0).to_list()
    np.save('/content/drive/MyDrive/Breast_Cancer/results/'+name+'/'+model+'.npy',data)

def get_histories_lists(historie):
    accur , val_accur = [] , []
    loss , val_loss = [] , []
    for hist in historie:
        accur.append(hist["accuracy"])
        val_accur.append(hist["val_accuracy"])
        loss.append(hist["loss"])
        val_loss.append(hist["val_loss"])
    return accur,val_accur,loss,val_loss

def get_avg_histories(model):
    names = ["accuracy","val_accuracy","loss","val_loss"]
    historie = load_histories(model)
    for name,item in zip(names,get_histories_lists(historie)):
        save_avg(item,name,model)

def get_vectors(folder):
    models = ["vgg16", "vgg19", "resNet50", "inceptionResNet","inceptionV3","xception","denseNet","mobileNet","Baseline"]
    to_return = {}
    for name in models:
        to_return[name] = np.load("/content/drive/MyDrive/Breast_Cancer/results/"+folder+"/"+name+".npy",allow_pickle=True)
    return to_return
```

```python
accuracies = get_vectors("val_accuracy")
plt.figure(figsize=(15,7))
models = ["vgg16", "vgg19", "resNet50", "inceptionResNet","inceptionV3","xception","denseNet","mobileNet","Baseline"]
for model , accu in accuracies.items():
    if model =="Baseline":
        plt.plot(range(200),accu,'r',linewidth=4)
    else :
        plt.plot(range(200),accu,linewidth=2.5)
plt.xlabel("Number of epochs",fontdict={"size":20})
plt.ylabel("Validation accuracy",fontdict={"size":20})
plt.legend(models,loc=0, prop={'size': 15})
plt.savefig('/content/drive/MyDrive/Breast_Cancer/test.png')
plt.show()
```
